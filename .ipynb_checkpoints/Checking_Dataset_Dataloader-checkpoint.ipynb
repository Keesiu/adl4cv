```python
import pandas as pd
import numpy as np
import torch
import matplotlib.pyplot as plt
from torchvision import transforms, utils
```

```python
from model import dataset
```

```python
label_frame = pd.read_csv('final_dataset/test.csv')

n = 2000
img_name = label_frame.iloc[n, 1]
labels = label_frame.iloc[n, 2]
labels = np.asarray(labels)

print('Image name: {}'.format(img_name))
print('Test label: {}'.format(labels))
```

```python
c256_dataset = dataset.Caltech_256_Dataset(csv_file='final_dataset\\train.csv',
                                   root_dir='final_dataset\\train_val')

fig = plt.figure()

for i in range(len(c256_dataset)):
    sample = c256_dataset[i]

    print(i, sample['image'].shape, sample['label'].shape)

    ax = plt.subplot(1, 4, i + 1)
    plt.tight_layout()
    ax.set_title('Sample #{} {}'.format(i, sample['label']))
    ax.axis('off')
    plt.imshow(sample['image'])
    
    if i == 3:
        plt.show()
        break
```

```python
cal_transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.RandomAffine(10),
    transforms.Resize((224,224)),
    transforms.RandomCrop((160,160)),
    transforms.Resize((224,224)),
    transforms.RandomHorizontalFlip(),
    #transforms.ToTensor(),
    #transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])


for i in range(len(c256_dataset)):
    sample = c256_dataset[i]

    ax = plt.subplot(2, 4, i + 1)
    plt.tight_layout()
    ax.set_title('Sample #{} {}'.format(i, sample['label']))
    ax.axis('off')
    plt.imshow(sample['image'])
    
    ax = plt.subplot(2, 4, i + 1 + 4)
    plt.tight_layout()
    ax.set_title('augmented')
    ax.axis('off')
    plt.imshow(cal_transform(sample['image']))
    
    if i == 3:
        plt.show()
        break
```

```python
transformed_dataset = dataset.Caltech_256_Dataset(csv_file='final_dataset\\train.csv',
                                          root_dir='final_dataset\\train_val',
                                          train_test = 'train')

for i in range(len(transformed_dataset)):
    sample = transformed_dataset[i]
    
    print(i, sample['image'].shape, sample['label'].shape)

    ax = plt.subplot(1, 4, i + 1)
    plt.tight_layout()
    ax.set_title('Sample #{} {}'.format(i, sample['label']))
    ax.axis('off')
    plt.imshow(sample['image'].permute(1, 2, 0).numpy())

    if i == 3:
        #print(sample['image'].permute(1, 2, 0).numpy())
        break
```

```python
dataloader = torch.utils.data.DataLoader(transformed_dataset,
                                             batch_size=4, shuffle=True,
                                             num_workers=4)

for i_batch, sample_batched in enumerate(dataloader):
    print(i_batch, sample_batched['image'].size(),
          sample_batched['label'].size())

    # observe 4th batch and stop.
    if i_batch == 3:
        plt.figure()
        grid = utils.make_grid(sample_batched['image'])
        plt.imshow(grid.numpy().transpose((1, 2, 0)))
        plt.axis('off')
        plt.ioff()
        plt.show()
        break
```
